```python
import numpy as np
import pandas as pd 
import seaborn as sns   
import matplotlib as mp

```

```python
data = pd.read_csv('../8_files/expense_data.csv')
data.head(1)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Month'] =  data['Date'].dt.month_name()
```
