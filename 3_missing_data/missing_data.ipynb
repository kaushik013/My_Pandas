# Missing data 

```python
import numpy as np
import pandas as pd 
```

```python
data = {
    'name' : [np.nan,'kaushik,','jay',np.nan,'mayur','pankaj',np.nan,'kajal'],
    'age' : [22,12,14,np.inf,np.nan,45,np.nan,25],
    'salary' : [20000,23000,120000,30000,np.inf,0,np.inf,80000],
    'dept' : ['IT',np.nan,'Account','Clerk','IT','Seles',np.nan,'IT']
}

df = pd.DataFrame(data)
df
```

```python
# define null value
df.isna()
```

```python
df.isna().any()
```

```python
# remove null value
df.dropna()
```

```python
# np.dropna(thresh = 3) ---> minimum 3 value not null honi chahiye
df.dropna(thresh=3)
```

# filling the data 

```python
df.fillna(0)
```

```python
df 
 
```

```python
values = {'name' : 'john', 'age' : 18,'salary' : 10000,'dept': 'clerk'}

df.fillna(value=values)
```
