# Series


```python
label = ['id','name','age','contact','address']
my_list = [1,'kaushik',22,9016883191,'Ahmedabad']
arr = np.array([11,12,13,14,15])
dic = {'a' : 100,'b':200, 'c' : 300, 'd' : 400, 'e' : 500}
```

```python
pd.Series(my_list,index=label)
```

```python
pd.Series(arr)
```

```python
pd.Series(dic)
```

```python
country = ['india','USA','UK','Pakistan','Shree lanka','Nepal']
pd.Series(country)
```

```python
pd.Series([11,12,13,14,15])
```

# custom index

```python
sub = ['English','Gujarati','Sanskrit','Math','Science']
marks = [99,79,89,78,88]

pd.Series(marks,index=sub)
```

```python
sub = ['English','Gujarati','Sanskrit','Math','Science']
marks = [99,79,89,78,88]

pd.Series(marks,index=sub,name="Kaushik's Marks")
```

# using dict


```python
my_data = {
    'English' : 88,
    'Hindi' : 90,
    'Gujarati' : 93,
    'Math' : 98,
    'Science' : 76
    }
```

```python
serise = pd.Series(my_data,name="kaushik's marks")
serise
```
