```python
import numpy as np
import pandas as pd 
```

```python
course = pd.read_csv('../8_files/courses.csv')
student = pd.read_csv('../8_files/students.csv')
nov = pd.read_csv('../8_files/reg-month1.csv')
dec = pd.read_csv('../8_files/reg-month2.csv')
matches = pd.read_csv('../8_files/matches.csv')
delevery = pd.read_csv('../8_files/deliveries.csv')
```

```python
regs = pd.concat([nov,dec],keys=['nov','dec'])
```

```python
# inner join 
student.merge(regs,how='inner',on='student_id')
```

```python
# left join 
course.merge(regs,how='left',on='course_id')
```

```python
# right join 
student.merge(regs,how='right',on='student_id')
```

```python
# outer join 
student.merge(regs,how='outer',on='student_id')
```

```python
# find total revenue generated 
regs.merge(course,how='inner',on='course_id')['price'].sum()
```

```python
# find month by month revenue 
temp = pd.concat([nov,dec],keys=['nov','dec']).reset_index()
```
