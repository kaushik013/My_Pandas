```python
ipl = pd.read_csv('../8_files/deliveries.csv')
```

```python
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman  with max no of six 
new = ipl[ipl['batsman_runs'] == 6]
new.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(1)
```

```python

# find the batsman with most number of 4'thand 6's in last 5 over 
new = ipl[(ipl['over'] > 15)]
temp = new[(new['batsman_runs'] == 4) | (new['batsman_runs'] == 6)]
temp.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(1)
```

```python
# find the virat kohli record against all team 
virat = ipl[ipl['batsman'] == 'V Kohli']
```

```python
grp = virat.groupby('bowling_team')
grp['batsman_runs'].sum().sort_values(ascending=False)
```

```python
# create a function that can return the highest score of any batsman
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
def find(name):
    new = ipl[ipl['batsman'] == name]
    return new.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head(1)

find('V Kohli')
```
