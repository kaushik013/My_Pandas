# agg method 

```python
grp.sum().head(1)
```

```python
grp.agg(
    { 
        'Runtime' : 'mean',
        'IMDB_Rating' : 'mean',
        'No_of_Votes' : 'sum',
        'Gross' : 'sum',
        'Metascore' : 'min'
    }
)
```

```python
df = pd.DataFrame(columns=movie.columns)
for obj,data in grp:
   temp = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
   df = pd.concat([df,temp])

df[['Series_Title','IMDB_Rating']]
```

```python
def check(g):
    return g['Series_Title'].str.startswith('A').sum()

grp.apply(check)
```

```python
def rank_app(gr):
    gr['ratting'] = gr['IMDB_Rating'].rank(ascending=False)
    return gr
```

```python
grp.apply(rank_app)
```

```python
def new_rat(group):
    group['noralization'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group


grp.apply(new_rat)
```
