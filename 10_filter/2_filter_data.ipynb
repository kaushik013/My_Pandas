# filter data 

```python
import numpy as np 
import pandas as pd 
```

```python
movie = pd.read_csv('../8_files/movies.csv')
```

```python
movie.head(10)
```

```python
# movie with ratting higher an 8 and votes > 10000
movie[(movie['imdb_rating'] > 8)  & (movie['imdb_votes'] > 10000)]['title_y'].count()
```

```python
# action movie with rating higher than 7.5 
movie[(movie['genres'].str.split('|').apply(lambda i : "Action" in i)) & (movie['imdb_rating'] > 7.5)]['genres'].count()
```
