# plot:
It is used to create a graph.

**Syntax**: plt.plot(x,y)

```python
x = np.arange(1,11)
y = x**2
x,y
```

```python
plt.plot(x,y , marker='v',color='brown',markerfacecolor='beige',markeredgecolor='black',label='square',markersize=10,linewidth=4)
plt.plot(x,x,label="numbers")
plt.xlabel('number')
plt.ylabel('square')
plt.grid(True)
plt.title('square of number')
plt.legend(loc='upper center',title='legend',fontsize=10,title_fontsize=20)

# for i,j in zip(x,y):
#   plt.annotate((int(i),int(j)),(i,j))

for i in range(len(x)):
  plt.annotate((int(x[i]),int(y[i])),(x[i],y[i]),va='center', ha='center')
plt.axvline(9,linestyle=':',color='black')
plt.axhline(90,linestyle=':',color='purple')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()

```

```python
df = pd.read_csv('/content/sales_data.csv',encoding='latin')
df
```

```python
total = df.pivot_table(values='Amount',index='Zone',aggfunc='sum').reset_index()
total
```

```python
plt.plot(total['Zone'],total['Amount'],marker='o',markerfacecolor='yellow',linestyle=" ")
# plt.annotate()
for i,j in zip(total['Zone'],total['Amount']):
  plt.annotate(j,(i,j))
plt.xlabel('Zone')
plt.ylabel('Total Sales')
plt.title('Zone vs Total Sales')
plt.axhline(41600873.45,color='red' ,linestyle=':',label='Higher')
plt.axvline('Central',linestyle=':',color='red')
plt.axhline(7049213.00,linestyle=':',color='green',label='lower')
plt.axvline('Eastern',color='green',linestyle=":")
plt.legend(loc='upper center',title='stat')
plt.grid()
plt.show()
```

```python
t = df.pivot_table(values='Amount',index='State',aggfunc='sum').reset_index()
t
```

```python
plt.plot(t['State'],t['Amount'],marker='s',color='red', markerfacecolor='blue' ,markeredgecolor='green',label='state sale')
plt.xticks(rotation=90)
for i,j in zip(t['State'],t['Amount']):
  plt.annotate(j//100000,(i,j),va='bottom',ha='right')
plt.grid()
plt.title('State vs Sale')
plt.legend()
plt.show()
```

# sub plots
**To create multiple plot in multiple graph we use sunplot.:**
Syntax: plt.subplot(no of rows, no of columns, position)

Position start with 1.

```python
x = np.arange(1,11)
y = np.arange(1,11)
y = x **2
z = x**3
x,y,z
```

```python
plt.plot(x,y)
plt.plot(x,z)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
plt.tight_layout()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,z)
plt.tight_layout()
```

# Figure
It is used to create a custom graph.

syntax: var = plt.figure()
        var1 = var.add_axes([left,bottom,width,height])
        var1.plot()
        var1.set_func()

```python
a = plt.figure()
b = a.add_axes([0,0,1,1])
plt.plot(x,y)
c = a.add_axes([0.5,0.1,0.4,0.3])
plt.plot(x,x**(1/3))
d = a.add_axes([0.08,0.65,0.4,0.3])
plt.plot(x,x**(1/4))
```

```python
a = plt.figure()
b = a.add_axes([0,0,1,1])
c = a.add_axes([0.5,0.1,0.4,0.3])
d = a.add_axes([0.08,0.65,0.4,0.3])
b.plot(x,y)
c.plot(x,x**(1/3))
d.plot(x,x**(1/4))

b.set_title('outer')
c.set_title('inner1')
d.set_title('inner2')

b.set_xlabel('numbers')
c.set_xlabel('numbers')
d.set_xlabel('numbers')
```
