```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
# 1. Number of people in a family
out = []
for i,j in zip(df['sibsp'],df['parch']):
  out.append(i+j+1)

```

```python
df['Total_Member'] = out
```

```python
df
```

```python
# 2. get the family name
out = []
for  i in df['name']:
  # i.split(',')
  out.append(i.split(',')[0])
```

```python
df['Family_Name'] = out
```

```python
df
```

```python
# 3. get the title of each person
out = []
for i in df['name']:
  t = i.split('.')
  out.append(t[0].split(',')[-1])
```

```python
df['title'] = out
```
