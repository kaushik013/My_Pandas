# Concatnation

The phenomenon of adding two or more data frames.

**Two types:**
*1. vertical* :
Syntax: pd.concat([df1,df2,...dfn])
this will go bottom by bottom.
*2. Horizontal* :
Syntax: pd.concat([df1,df2,....dfn],axis=1)




```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

# Merge/Join

The phenomenon  of joining 2 dataframes.

**Types:**
*1. inner join :*
Inner join is the phenomenon of get the matched record of both data frame
Syntax: pd.merge(df1,df2,on = 'commmon_col',how='inner')
for innner join you dont need to mention how it will take it automatically


*2. left join :*:
Left join is the phenomenon of get the matched record of both data frame and unmatched data from left dataframe
Syntax: pd.merge(df1,df2,on = 'commmon_col',how='left')


*3. right join :*
Right join is the phenomenon of get the matched record of both data frame and unmatched data from Right dataframe
Syntax: pd.merge(df1,df2,on = 'commmon_col',how='right')


*4. outer join :*
Outer join is the phenomenon of get the matched data of both data frame and unmatched data of both data frame.
Synax: pd.merge(df1,df2,on = 'commmon_col',how='outer')



```python
pd.merge(df1,df2,on='deptno',how='inner')
```

```python
pd.merge(df1,df2,on='deptno',how='left')
```

```python
pd.merge(df1,df2,on='deptno',how='right')
```

```python
pd.merge(df1,df2,on='deptno',how='outer ')
```
