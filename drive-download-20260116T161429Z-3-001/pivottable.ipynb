# Pivot table

When we are going to use multiple aggregate function on multiple columns.

**Note:**
aggfunc:
This are the inbuilt function we are going to perform.

index:
Column_name by which you are going to categorise.

values:
Columns on which you are going to perform aggregate function.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
a = df.groupby('embarked')
print('minimum age in S:',a.get_group('S')['age'].min())
print('maximum age in S:',a.get_group('S')['age'].max())
print('minimum age in C:',a.get_group('C')['age'].min())
print('maximum age in C:',a.get_group('C')['age'].max())
print('minimum age in Q:',a.get_group('Q')['age'].min())
print('maximum age in Q:',a.get_group('Q')['age'].max())

```

```python
df.pivot_table(values='age',index='embarked',aggfunc=['min','max'])
```

```python
df.pivot_table(values='age',index='survived',aggfunc=['min','max'])
```

```python
df.pivot_table(values='age',index=['survived','sex'],aggfunc=['min','max'],)
```

```python
df.pivot_table(values='fare',index='sex',aggfunc=['mean','median'])
```

```python
df = pd.read_csv('sales_data.csv',encoding='latin')
df
```

```python
# 1. Find the average sales based on state

print('average sales based on state:')
print(df.pivot_table(values='Amount',index='State',aggfunc=['mean']))
```

```python
# 2. find the maximum amount spent by male and female
print('maximum amount spent by male and female:')
print(df.pivot_table(values='Amount',index='Gender',aggfunc='max'))
```

```python
# 3. find the maximum number of orders placed by zone
print('maximum number of orders placed by zone:')
print(df.pivot_table(values='Orders',index='Zone',aggfunc='max'))
```

```python
# 4. Find the median amount spend by each occupation
print('median amount spend by each occupation:')
print(df.pivot_table(values='Amount',index='Occupation',aggfunc='median'))
```

```python
# 5. Find the age group which spent the most amount of money.
print('age group which spent the most amount of money:')
print(df.pivot_table(values='Amount',index='Age Group',aggfunc='sum'))
```

```python
# 6. Find maritial status who ordered more
print('maritial status who ordered more:')
print(df.pivot_table(values='Orders', index='Marital_Status',aggfunc='sum'))
```
