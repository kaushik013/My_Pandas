# Date Range

It is function which is used to generate a sequence of date/time.

Syntax: pd.date_range(start = 'mm/dd/yyyy', end='mm/dd/yyyy')

freq : It is used to determine the gap between 2 data point.

Freq:

YS-Year start
YE-Year End
YE-Mon - It determined from where to count year end based on month
YS-Mon - It determined from where to count year start based on month
MS-month start
ME-month end
W - week
W-DAY - It determined on which day week will be start
D - day
h - hour
min - minutes
s - second


```python
import numpy as np
import pandas as pd
```

```python
pd.date_range(start='12/1/2025',end='31/12/2025')
```

```python
pd.date_range(start='11/21/2005',end='12/25/2025',freq='min')
```

# to_datetime

It is used to convert a string column into date time column.

Syntax: pd.to_datetime('col')

```python
df= pd.read_csv('/content/amazon_fires.csv',encoding='latin')
df
```

```python
df.info()
```

```python
df['encontro'] = pd.to_datetime(df['encontro'])
```

```python
df.info()
```

```python
df['year'] = df['encontro'].dt.year
```

```python
df
```

```python
df['month'] = df['encontro'].dt.month
```
