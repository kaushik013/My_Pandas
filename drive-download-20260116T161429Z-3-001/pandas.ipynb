```python
prime = []
for i in range(2,101):
  for j in range(2,i):
    if i%j==0:
      break
  else:
    prime.append(i)
# print(prime)

series = pd.Series(prime,name='Prime numbers')
series
```

```python
def isprime(n):
  for i in range(2,n):
    if n%i == 0:
      return False
  return True
var = list(filter(isprime,range(2,101)))
prime = pd.Series(var,name='Prime numbers')
prime
```
