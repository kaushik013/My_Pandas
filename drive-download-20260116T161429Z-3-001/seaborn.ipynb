# 2. Categorical plot


* It is used to analyze and cvisualize categorical column.
* Types of categorical plot:

1. count plot:
* It is used to count number of values present a unique value in each and every column
* Syntax: sns.countplot(data=df,x='',y='')

2. barplot:
* It is going to display average of a numerical column with respect to categorical column.
* syntax: sns.barplot(data=df,x='For making look good categorical',y='numerical')

3. scatter plot:
* It is used to plot every value of x with respective y.
* Syntax: sns.scatterplot(data=df,x='',y='')

4. violin plot:
* It is used to find most populated area of a particular column.
* syntax:sns.violinplot(data=df,x='categorical column is recomended',y='numerical column')


5. box plot:
* It is used to find the outliers,
* Syntax: sns.boxplot(data=df,x='',y='')

```python
# steps for finding outlier
q1 =  df['pH'].quantile(0.25)
q3 = df['pH'].quantile(0.75)
iqr = q3-q1
# iqr
ul=q3+(iqr*1.5)
ll = q1-(iqr*1.5)
ul,ll
```

```python
# get outlier
df[(df['pH']>ul)| (df['pH']<ll)]
```

```python
df1 = df[(df['pH']<ul) & (df['pH']>ll)]
sns.boxplot(data=df1,y='pH')
```

```python
# Winsorization // Capping
# The Phenomenon of convertong outliers above the upper limit into upper limit and outliers below your lower limit to lower limit.
out = []
for i in df['pH']:
  if i>ul:
    out+=[ul]
  elif i<ll:
    out+=[ll]
  else:
    out+=[i]
df['pH'] = out
```

```python
sns.boxplot(data=df,y='pH')
```

# 3. Matrix plot:
* It is used to understand behaviour of one column with respect to another column.

* Types of Matrix plot:
1. heatmap:
* Syantax: sns.heatmap(data = df, annot = True,cmap = )
* Two ways to get data:
1. pivot_table()

2. corr():
* It is used to find the relation.
* Syntax df.corr()

```python
df.columns
```

```python
a = df.pivot_table(values =['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol'], index = 'quality',aggfunc='sum' )
a
```

```python
plt.figure(figsize=(15,3))
sns.heatmap(data = a,annot=True,cmap='viridis')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data=a,annot=True)
```
