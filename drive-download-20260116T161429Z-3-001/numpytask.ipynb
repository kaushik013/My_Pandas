```python
import numpy as np
```

```python
# A 1D array of 10 equally spaced values between 0 and 50.
arr1 = np.linspace(0,50,10)
arr1
```

```python
# A 2D array of shape (3x3) containing random integers between 5 and 20.
arr2 = np.random.randint(5,20,9).reshape(3,3)
arr2
```

```python
# Print the dimensions of both arrays.
print("Dimension of array1 is: ",arr1.ndim)
print("Dimension of array2 is: ",arr2.ndim)
```

```python
# Print the shape of both arrays.
print("Shape of array1 is: ",arr1.shape)
print("Shape of array2 is: ",arr2.shape)
```

```python
# Reshape the 1D array into a 2D array with shape (2x5).
arr1 = arr1.reshape(2,5)
# Print the reshaped array.
print(arr1)
# Flatten the reshaped array back to 1D and print it
arr1 = arr1.reshape(-1)
print("Flatten array is: ",arr1)
```

```python
# For the 2D array created in Part 1, print the element present at the 2nd row and 3rd column.
print(arr2[1,2])
# Extract the first two rows and the last two columns of the array.
print(arr2[0:2,1:])
```

```python
# Slice the first four elements from the flattened 1D array
print(arr1[0:4])
# Reverse the elements of the 1D array using slicing.
print(arr1[::-1])
```

```python
# Create another 1D array of 10 integers between 1 and 10
arr3 = np.arange(1,11)
# Perform the following element-wise operations between the 1D arrays created in Part 1 and Part 3:
# Addition
print(arr1+arr3)
# Multiplication
print(arr1*arr3)
# Subtract a constant value (e.g., 5) from the array.
print(arr1/5)

```

```python
# Create a copy of the original 2D array from Part 1.
arr6 = np.copy(arr2)
arr6
# Modify one element in the copy and show that it doesnâ€™t affect the original array.
arr6[1][1] =10
print(arr2)
print(arr6)

# Create a view of the original 2D array and modify one element. Show that it does affect the original array.
arr7 = arr2.view()
arr7[0,2]=165
print(arr2)
print(arr7)

# In summary copy will not  affect the original array but view will affect
```
