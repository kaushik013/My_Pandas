# Feature Engineering
The phenomenon of creating a new features with the help of old data and existing data
Syntax: df['new_Col_name'] = value

```python
import numpy as np
import pandas as pd
```

```python
d = {
    'name':['rohit','dinesh','anand','mitesh','anas','hardik','abhishek'],
    'roll no':[1,2,3,4,5,6,7],
    'english':[72,97,55,67,100,82,70],
    'math':[69,45,60,60,100,87,58],
    'science':[65,70,50,88,100,88,78],
    'gujarati':[10,0,70,73,100,79,56]
}
```

```python
df = pd.DataFrame(d)
df
```

```python
df['total'] = df['english']+df['math']+df['science']+df['gujarati']
df
```

```python
df['percentage'] = df['total']/4
df
```

```python
out = []
for i,j,k,l in zip(df['english'],df['math'],df['science'],df['gujarati']):
  if i>=35 and j>=35 and k>=35 and l>=35:
    out.append('pass')
  else:
    out.append('fail')
```

```python
df['result'] = out
df
```

```python
out =[]
for i,j in zip(df['percentage'],df['result']):
  if 35<i<=49 and j == 'pass':
    out.append('D')
  elif 50<i<=69 and j == 'pass':
    out.append('C')
  elif 70<i<=84 and j == 'pass':
    out.append('B')
  elif 85<i<=100 and j == 'pass':
    out.append('A')
  else:
    out.append('fail')
```

```python
df['grade'] = out
df
```
