# Inbuilt functions of pandas for for dataframes

1. head():
By deafult it will reaturn top 5 rown and by passing any integer number(n) it will return n rows from top.

2. tail():
By default it will return last 5 rows from data frame and if we pass integer n it return n number of rows.

3. sample():
By default it will return a single row randomly but if we pass integer n it will return n number of random rows.

4. info():
It is used to get the inforamation about the dataframe.

5. describe():
It is used to get statastical data of numerical column by default and for object data type we are using df.describe(include = object).

6. unique():
Syntax: df['column_name'].unique()
It is used to find the unique values present in the particular column.

7. value_counts():
Syntax: df['column_name'].value_counts()
It is used to count a ocurence of unique values.

8. sort_values('column_name'):
syntax: df.sort_values('column_name')
It is used to sort a particular column in ascending order by default but for descending we need to use attribute called as asscending = False

9. isnull():
It is used to find and check whether the column is containing NaN value or not.

10. notnull():
It is used to extract all the value expect null.
