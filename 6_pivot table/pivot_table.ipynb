# pivot table 

```python
import numpy as np 
import pandas as pd 
```

```python

data = {
    'Emp_ID': range(1, 21),
    'Name': [
        'Amit','Rahul','Neha','Priya','Rohan',
        'Kiran','Sneha','Arjun','Pooja','Vikas',
        'Anita','Suresh','Meena','Ravi','Kavita',
        'Nikhil','Payal','Deepak','Isha','Manoj'
    ],
    'Department': [
        'IT','HR','IT','Finance','HR',
        'IT','Finance','IT','HR','Finance',
        'IT','HR','Finance','IT','HR',
        'Finance','IT','HR','Finance','IT'
    ],
    'City': [
        'Delhi','Mumbai','Delhi','Pune','Mumbai',
        'Delhi','Pune','Mumbai','Delhi','Pune',
        'Mumbai','Delhi','Pune','Delhi','Mumbai',
        'Delhi','Pune','Mumbai','Delhi','Pune'
    ],
    'Salary': [
        50000,42000,60000,70000,45000,
        65000,72000,58000,48000,75000,
        62000,46000,71000,64000,47000,
        73000,66000,49000,76000,68000
    ],
    'Experience': [
        2,3,5,7,4,
        6,8,5,3,9,
        6,4,8,5,3,
        9,6,4,10,7
    ]
}

df = pd.DataFrame(data)
df
```

```python
pd.pivot_table(df,values='Salary',index='City',columns='Department',aggfunc='max')
```
