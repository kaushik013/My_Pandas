# concatination


```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'A1' : ['A1','A2','A3'],
    'B1' : ['B1','B2','B3'],
    'C1' : ['C1','C2','C3']
})

df2 = pd.DataFrame({
    'A1' : ['X1','X2','X3'],
    'B1' : ['Y1','Y2','Y3'],
    'C1' : ['Z1','Z2','Z3']
 })
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```
