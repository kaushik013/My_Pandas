# Merging

```python
import numpy as np 
import pandas as pd 
```

```python
employee = pd.DataFrame({
    'Emp_id' : [1,2,3,4,5],
    'name' : ['john','jay','kartik','sanjay','yug'],
    'department' : ['HR','IT','Finance','IT','HR'],
})

salary = pd.DataFrame({
    'Emp_id' : [1,2,3,6,7],
    'salary' : [10000,30000,50000,30000,10000]
}) 
```

```python
employee
```

```python
salary
```

```python
# only merge only of the same employee id's
pd.merge(employee,salary)
```

```python
pd.merge(employee,salary,on='Emp_id')
```

# inner

```python
pd.merge(employee,salary,on='Emp_id',how='inner')
```

# outer

```python
pd.merge(employee,salary,on='Emp_id',how='outer')
```

# left


```python
pd.merge(employee,salary,on='Emp_id',how='left')
```

# right


```python
pd.merge(employee,salary,on='Emp_id',how='right')
```
