# joining


```python
import numpy as np 
import pandas as pd 
```

```python
df1 = pd.DataFrame({
    'name' : ['kaushik','Raj','Pankaj']
},index=[1,2,3])

df2 = pd.DataFrame({
    'Score' : [10,2,13]
},index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
# inner
df1.join(df2,how='inner')
```

```python
# outer
df1.join(df2,how='outer')
```

```python
# left
df1.join(df2,how='left')
```

```python
# right
df1.join(df2,how='right')
```
