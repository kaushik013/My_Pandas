```python
ipl = pd.read_csv('../8_files/ipl-matches.csv')
```

```python
ipl.head(1)
```

```python
ipl.value_counts()
```

```python
# which player has won most potm -> in final and qualifier
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```
