```python
import numpy as np 
import pandas as pd 
```

```python

data = {
    'name': [
        'nitish', 'ankit', 'rupesh', np.nan, 'mrityunjay',
        np.nan, 'rishabh', np.nan, 'aditya', np.nan
    ],
    'college': [
        'bit', 'iit', 'vit', np.nan, np.nan,
        'visi', 'ssit', np.nan, np.nan, 'git'
    ],
    'branch': [
        'eee', 'it', 'cse', np.nan, 'me',
        'ce', 'civil', 'cse', 'bio', np.nan
    ],
    'cgpa': [
        6.66, 8.25, 6.41, np.nan, 5.6,
        9.0, 7.4, 10.0, np.nan, np.nan
    ],
    'package': [
        4, 5, 6, np.nan, 6,
        7, 8, 9, np.nan, np.nan
    ]
}

student = pd.DataFrame(data)
student
```

# dropna --> series &  dataframe 

```python
student['name'].dropna()
```

```python
student.dropna()
```

```python
# only if all data is nan 
student.dropna(how='all')
```

```python
# only if we want to delate perticular row
student.dropna(subset=['name','cgpa'])
```

# fillna --> add value 

```python
student['name'].fillna('unknown')
```

```python
student.fillna(0)
```

```python
student['package'].fillna(student['package'].mean())
```

# forwardfill --> ffill & backwardfill --> bfill 

```python
student['name'].ffill()
```

```python
student['name'].bfill()
```

```python
ipl = pd.read_csv('../8_files/ipl-matches.csv')
```

```python
# find the last match played by virat kohli in delhi 
ipl.head(1)
```

```python
ipl['all_player'] = ipl['Team1Players'] + ipl['Team2Players']
```

```python
ipl.head(1)
```

```python
def chech_ply(player):
    return 'V Kohli' in player
```

```python
ipl['play_kohli'] = ipl['all_player'].apply(chech_ply)
```

```python
ipl.head(1)
```

```python
ipl[(ipl['play_kohli'] == True) & (ipl['City'] == 'Delhi')].sort_values('Date',ascending=False).drop_duplicates(subset=['City','play_kohli'],keep='first')
```
