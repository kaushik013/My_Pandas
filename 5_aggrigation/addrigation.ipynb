```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame({
    'category' : ['A','B','A','B','A','B','A','B',],
    'Store' : ['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales' : [100,300,200,430,200,100,30,140],
    'Quantity' : [10,15,12,13,10,13,14,15],
    'date' : pd.date_range('2025-01-01',periods=8)
})
data
```

```python
# sum of sales group by
data.groupby('category')['Sales'].sum()
```

```python
data.groupby('Store')['Sales'].sum()
```

```python
# multiple category
data.groupby(['category','Store'])['Sales'].sum()
```
