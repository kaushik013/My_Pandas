# aggrigation function

```python
import numpy as np 
import pandas as pd 
```

```python
data = pd.DataFrame({
    'category' : ['A','B','A','B','A','B','A','B',],
    'Store' : ['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales' : [100,300,200,430,200,100,30,140],
    'Quantity' : [10,15,12,13,10,13,14,15],
    'date' : pd.date_range('2025-01-01',periods=8)
})
data
```

```python
# mean 
data['Sales'].mean()
```

```python
# median ---> Data ko ascending order me sort karta hai Middle value return karta hai
data['Sales'].median()
```

```python
# mode() ---> function Series ya DataFrame me se most frequent value(s) nikalta hai.
data['Sales'].mode()
```

```python
# max 
data['Sales'].max()
```

```python
# min 
data['Sales'].min()
```

```python
# count 
data['Sales'].count()
```

```python
# std 
data['Sales'].std()
```

```python
# multiple 
data['Sales'].agg(['sum','max','min','mean'])
```
