```python
import numpy as np 
import pandas as pd 
import random
```

```python


np.random.seed(7)

rows = 250

names = [
    "Amit Sharma", "Rahul Verma", "Neha Patel", "Priya Mehta", "Sanjay Singh",
    "Ravi Kumar", "Kiran Joshi", "Anita Desai", "Vijay Shah", "Pooja Nair",
    "Ramesh Iyer", "Seema Malhotra", "Ajay Gupta", "Nisha Kapoor", "Deepak Yadav",
    "Manoj Mishra", "Sneha Kulkarni", "Arjun Reddy", "Swati Chavan", "Rohit Jain",
    "Alok Tiwari", "Mehul Shah", "Payal Joshi", "Vikas Agarwal", "Komal Patel",
    "Harsh Mehta", "Divya Bansal", "Suresh Pawar", "Kavita Kulkarni", "Nitin More"
]

departments = ["IT", "HR", "Finance", "Sales", "Marketing"]
cities = ["Delhi", "Mumbai", "Pune", "Bangalore", "Chennai", "Hyderabad"]

df = pd.DataFrame({
    "emp_id": range(1, rows + 1),
    "name": [random.choice(names) for _ in range(rows)],
    "age": np.random.randint(21, 60, rows),
    "salary": np.random.randint(25000, 120000, rows),
    "department": np.random.choice(departments, rows),
    "city": np.random.choice(cities, rows),
    "experience": np.random.randint(1, 11, rows),
    "rating": np.round(np.random.uniform(1, 5, rows), 1),
    "gender": np.random.choice(["Male", "Female"], rows)
})

df.head()
```

# Tasks

```python
# Display the last 10 records of the DataFrame.
df[-10:]
```

```python
# Display all employees whose age is greater than 35.
df[df['age'] > 35]
```

```python
# Display all employees from the IT department.
df[df['department'] == 'IT']
```

```python
# Display female employees from Mumbai.
df[(df['gender'] == 'Female') & (df['city'] == 'Mumbai')]
```

```python
# Display employees whose salary is between 50,000 and 80,000. 
df[(df['salary'] >= 50000) & (df['salary'] <= 80000)]
```

```python
# Find the average salary department-wise.
df.groupby('department')['salary'].mean()
```

```python
# Find the maximum salary city-wise.
df.groupby('city')['salary'].max()
```

```python
# Find the total salary per department. 
df.groupby('department')['salary'].sum()
```

```python
# Count employees gender-wise.
df.groupby('gender').size()
```

```python
# Count missing values in each column.
df.isna().sum()
```

```python
# Display the top 10 highest paid employees.
df.sort_values('salary',ascending=False).head(10)
```

```python
# Display the bottom 5 lowest ratings.
df.sort_values('rating').head(5)
```

```python
# Sort employees by department and then by age (ascending).
df.sort_values(['department','age'])
```

```python
# Display employees whose salary is greater than the department average salary.
df[df['salary'] > df.groupby('department')['salary'].transform('mean')]

```
