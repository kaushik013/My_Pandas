```python
# how many countries are there whoes political leaders are unknown
data[data['political_leader'].isna()]['country'].count()
```

```python
count = 0 
def copunt_rep(text):
    global count
    if('republic' in text.lower()):
        count += 1
    return text

data['country_long'] = data['country_long'].apply(copunt_rep)
count
```
