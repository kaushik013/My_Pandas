```python
# the soluction ---> multi index series (also known as hierarchical indexing)
# multiple index level within a single index 
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex  = pd.MultiIndex.from_tuples(index_val)
multiindex
```

```python
multiindex.levels[0]
```

```python
# multiindex from tuple 
multi = pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
multi
```

```python
# create series with multipleindex  object
s = pd.Series([1,2,3,4,5,6,7,8],index=multi)

```

```python
# how to fatch a data 
s
```

```python
s['cse'][2021]
```
